**Data strcutures and Algorithms**

Given an array/list(ARR) of length N, you need to find and return the sum of all the elements in the array/list.

```python
def sum(arr,N):
  sum=0
  for i in range(N):
    sum=sum+arr[i]

  return sum

sum([1,2,3,4],4)
```

Find the largest element in the arr/list

```python
def find_largest(arr):
  largest=arr[0]
  for i in range(1,len(arr)):
    if(arr[i]>largest):
      largest=arr[i]

  return largest

find_largest([1,2,3,4,5])
```

Find the minimum element in a list/arr

```python
def find_minimum(arr):
  smallest=arr[0]

  for i in range(1,len(arr)):
    if(arr[i]<smallest):
      smallest=arr[i]

  return smallest

find_minimum([1,2,3,4,5])
```

Linear Search

```python
def linear_search(arr,search_element):
  for i in range(len(arr)):
    if(arr[i]==search_element):
      return i

linear_search([1,2,3,4],4)
```

Sum of two arrays


```python
def sum_of_two_arrays(arr1,arr2):
  for i in range(len(arr1)):
    arr1[i]=arr1[i]+arr2[i]

  return arr1

ans=sum_of_two_arrays([1,2,3],[4,5,6])
print(ans)

```

Two Sum

```python
def two_sum(arr,target):
  arr=sorted(arr)
  left=0
  right=len(arr)-1
  count=0

  for i in range(len(arr)):
    if(arr[left]+arr[right]==target):
      count=count+1
      left=left+1
      right=right-1
    elif(arr[left]+arr[right]<target):
      left=left+1
    else:
      right=right-1
  print(count)

two_sum([2 ,8, 10, 5, -2, 5],10)

```

Swap Alternate

```python
def swap(arr):
  if(len(arr)<2):
      print("atleast two elements are required")\

  i=0
  j=1

  while(j<len(arr)):


    temp=arr[i]
    arr[i]=arr[j]
    arr[j]=temp

    i+=2
    j+=2

  print(arr)

swap([1,2,3,4,5,6,7])



```

Arrange numbers as follows

input:1,6


output:
1 3 5 6 4 2

```python
def arrange(num1, num2):

    total_elements = num2 - num1 + 1
    arr = [0 for _ in range(total_elements)]

    left = 0
    right = len(arr) - 1
    current_element = num1


    while left <= right:
        if left == right:
            arr[left] = current_element
            break

        arr[left] = current_element
        current_element += 1

        arr[right] = current_element
        current_element += 1

        left += 1
        right -= 1

    print(arr)

arrange(1, 6)

```

Maximum candies


```python
def max_candies(arr,extra_candies):
  max_candy=max(arr)
  ans=[]
  for i in range(0,len(arr)):
    arr[i]+=extra_candies
    if(arr[i]>max_candy):
      ans.append(True)
    else:
      ans.append(False)

  return ans

ans=max_candies([2,3,5,1,3],3)
print(ans)



```

Find the unique element in the array

```python
def find_unique(arr):
    unique_element = None
    for num in arr:
        if arr.count(num) == 1:
            unique_element = num
            break
    return unique_element
unique=find_unique([1,2,1,2,3])
print(unique)
```

Intersection of Two arrays

```python
def intersection(arr1,arr2):
  ans=[]
  for i in range(len(arr1)):
    for j in range(len(arr2)):
      if(arr1[i]==arr2[j]):
        ans.append(arr1[i])
        arr2[j]=-1
  print(ans)

intersection([1,2,3,4],[2,3,4,5])

```
